# CRIAÇÃO DATASETS

- CODIFICAÇÃO 1-out-of-C

- Setosa = [1,0,0]
- Versicolor = [0,1,0]
- Virginica = [0,0,1]

```python
import pandas as pd
import numpy as np
```

```python
dataset_setosa = pd.read_csv("datasets/iris/df_setosa", sep=',')
dataset_versicolor = pd.read_csv("datasets/iris/df_versicolor", sep=',')
dataset_virginica = pd.read_csv("datasets/iris/df_virginica", sep=',')
```

```python
d1_setosa = []
d2_setosa = []
d3_setosa = []
for lin in range(0, len(dataset_setosa)):
    if dataset_setosa["class"][lin] == 1.0:
        d1_setosa.append(1)
        d2_setosa.append(0)
        d3_setosa.append(0)
    else:
        d1_setosa.append(0)
        d2_setosa.append(0)
        d3_setosa.append(0)
        
dataset_setosa["d1"] = d1_setosa
dataset_setosa["d2"] = d2_setosa
dataset_setosa["d3"] = d3_setosa
df_setosa_d = dataset_setosa.drop('class', axis=1, inplace=False)
```

```python
d1_versicolor = []
d2_versicolor = []
d3_versicolor = []
for lin in range(0, len(dataset_versicolor)):
    if dataset_versicolor["class"][lin] == 1.0:
        d1_versicolor.append(0)
        d2_versicolor.append(1)
        d3_versicolor.append(0)
    else:
        d1_versicolor.append(0)
        d2_versicolor.append(0)
        d3_versicolor.append(0)
        
dataset_versicolor["d1"] = d1_versicolor
dataset_versicolor["d2"] = d2_versicolor
dataset_versicolor["d3"] = d3_versicolor
df_versicolor_d = dataset_versicolor.drop('class', axis=1, inplace=False)
```

```python
d1_virginica = []
d2_virginica = []
d3_virginica = []
for lin in range(0, len(dataset_virginica)):
    if dataset_versicolor["class"][lin] == 1.0:
        d1_virginica.append(0)
        d2_virginica.append(0)
        d3_virginica.append(1)
    else:
        d1_virginica.append(0)
        d2_virginica.append(0)
        d3_virginica.append(0)
        
dataset_virginica["d1"] = d1_virginica
dataset_virginica["d2"] = d2_virginica
dataset_virginica["d3"] = d3_virginica
df_virginica_d = dataset_virginica.drop('class', axis=1, inplace=False)
```

```python
df_setosa_d.to_csv('datasets/iris/df_setosa_d', sep=',', index=False)
df_versicolor_d.to_csv('datasets/iris/df_versicolor_d', sep=',', index=False)
df_virginica_d.to_csv('datasets/iris/df_virginica_d', sep=',', index=False)
```

# ARTIFICIAL I

```python
import random
from matplotlib import pyplot as plt
```

```python
x1_c = []
x2_c = []

x1_e = []
x2_e = []

x1_t = []
x2_t = []

circulo = []
estrela = []
triangulo = []

for i in range(50):
    x1c = (random.uniform(0.1, 0.3))
    x1_c.append(x1c)
                
    x2c = (random.uniform(0.6, 0.9))
    x2_c.append(x2c)
    
    x1e = (random.uniform(0.35, 0.6))
    x1_e.append(x1e)
           
    x2e = (random.uniform(0.25, 0.5))
    x2_e.append(x2e)
           
    x1t = (random.uniform(0.65, 0.9))
    x1_t.append(x1t)
           
    x2t = (random.uniform(0.65, 0.9))
    x2_t.append(x2t)
           
    circulo.append([x1c, x2c,1, 0, 0])
    estrela.append([x1e, x2e,0, 1, 0])
    triangulo.append([x1t, x2t,0, 0, 1])
```

```python
columns = ["x1", "x2", "d1", "d2", "d3"]
```

```python
df_circulo = pd.DataFrame(data=circulo, columns=columns)
df_estrela = pd.DataFrame(data=estrela, columns=columns)
df_triangulo = pd.DataFrame(data=triangulo, columns=columns)
```

```python
df_circulo.to_csv('datasets/artificial_1/df_circulo', sep=',', index=False)
df_estrela.to_csv('datasets/artificial_1/df_estrela', sep=',', index=False)
df_triangulo.to_csv('datasets/artificial_1/df_triangulo', sep=',', index=False)
```

```python
import numpy as np
from sklearn import datasets
iris = datasets.load_iris()
```

```python
import pandas as pd
df = pd.DataFrame(data=iris['data'], columns=iris.feature_names)
```

```python
df.head()
```

```python
iris_labels = pd.DataFrame(data=iris["target"], columns=["label"])
iris_labels.head()
```

```python
df_iris = pd.concat([df, iris_labels], axis=1)
```

```python
df_iris.head()
```

```python
d_set = []
d_vers = []
d_virg = []
for lin in range(0, len(df_iris)):
    if df_iris["label"][lin] == 0:
        d_set.append(1)
        d_vers.append(0)
        d_virg.append(0)
    elif df_iris["label"][lin] == 1:
        d_set.append(0)
        d_vers.append(1)
        d_virg.append(0)
    elif df_iris["label"][lin] == 2:
        d_set.append(0)
        d_vers.append(0)
        d_virg.append(1)

df_iris["d_set"] = d_set
df_iris["d_vers"] = d_vers
df_iris["d_virg"] = d_virg
df_iris_d = df_iris.drop('label', axis=1, inplace=False)
```

```python
df_iris_d.to_csv('datasets/iris/df_iris', sep=',', index=False)
```
